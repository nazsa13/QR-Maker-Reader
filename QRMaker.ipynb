```python
# Run This Block To Generate A QR
import qrcode
import os
from IPython.display import Image, display

# Ask user for link/text and file name
data = input("Enter the link or text for the QR code: ")
filename = input("Enter the file name (without extension): ") + ".png"

# Generate QR code
qr = qrcode.QRCode(version=1, box_size=10, border=5)
qr.add_data(data)
qr.make(fit=True)

img = qr.make_image(fill="black", back_color="white")

# Save in current working directory
save_path = os.path.join(os.getcwd(), filename)
img.save(save_path)

print(f"✅ QR code saved as: {save_path}")

# Show QR code directly in the notebook
display(Image(filename=save_path))
```

```python
# Run This Block To Reveal The QR Content

import cv2
from pyzbar.pyzbar import decode
from IPython.display import display, Image
import os

# Ask for file name
filename = input("Enter the QR code image filename (with extension): ")

# Check if file exists
if not os.path.exists(filename):
    print(f"❌ Error: The file '{filename}' was not found. Please check the name and try again.")
else:
    # Load the image
    img = cv2.imread(filename)

    # Decode the QR code
    decoded_objects = decode(img)

    if decoded_objects:
        for obj in decoded_objects:
            print("✅ QR Code Data:", obj.data.decode("utf-8"))
    else:
        print("⚠️ No QR code found in the image.")

    # Show the QR code image in notebook
    display(Image(filename=filename))
```
